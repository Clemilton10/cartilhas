## ``EXEMPLO 1``
```
Objetivo: Calcular o imposto de cada preco
```

````python
precos = [ 1000, 1500, 1250, 2500 ]

def adicionar_imposto(preco):
    return preco * 1.1

# do jeito normal
# precos_com_imposto = []
# for preco in precos:
#    precos_com_imposto.append(adicionar_imposto(preco))

# com o MAP
# assim retorna um objeto MAP
# precos_com_imposto2 = map( adicionar_imposto, precos )

# modo completo
precos_com_imposto3 = list( map( adicionar_imposto, precos ) )
print(precos_com_imposto3)
````

## ``EXEMPLO 2``
```
Objetivo: Padronizar os itens da lista
```

````python
def padronizar_texto(texto):
    texto = texto.casefold()
    texto = texto.replace("  ", " ")
    texto = texto.strip()
    return texto

produtos = [' ABC12 ', 'abc34', 'AbC37', 'beb12', ' BSA151', 'BEB23']

# jeito normal
# for i, produto in enumerate( produtos ):
#    produtos[i] = padronizar_texto( produto )

produtos = list( map( padronizar_texto, produtos ) )
print( produtos )
````
