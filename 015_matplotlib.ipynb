```python
vendas_meses = [1500, 1727, 1350, 999, 1050, 1027, 1022, 1500, 2000, 2362, 2100, 2762]
meses = ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 4))
plt.plot(meses, vendas_meses)
plt.ylabel('Vendas')
plt.xlabel('Meses')
# max -> maior número da lista
# x_inicio, x_final, y_inicio, y_final
plt.axis( [0, 12, 0, max( vendas_meses ) + 100 ] )
plt.show()
```

```python
import numpy as np
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)

plt.figure(figsize=(18, 4))
plt.plot(meses, vendas, '--', linewidth=4.0, color='#ff8800')
plt.ylabel('Vendas')
plt.xlabel('Meses')
# max -> maior número da lista
# x_inicio, x_final, y_inicio, y_final
plt.axis( [0, 50, 0, max( vendas_meses ) + 500 ] )
plt.show()
```

```python
# dimensao do gráfico
plt.figure(figsize=(18, 4))

# primeiro posição (linha, q_graficos_horizontais, posicao_grafico)
plt.subplot(1, 3, 1)

# gráfico de barra
plt.bar(meses, vendas, color='red')

# segunda posição (linha, q_graficos_horizontais, posicao_grafico)
plt.subplot(1, 3, 2)

# gráfico de bolinhas
plt.scatter(meses, vendas, color='green')

# terceira posição (linha, q_graficos_horizontais, posicao_grafico)
plt.subplot(1, 3, 3)

# gráfico de linhas
plt.plot(meses, vendas, color='blue')

# título
plt.suptitle('Categorical Plotting')

plt.show()
```
